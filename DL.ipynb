날씨 예측에 필요한 라이브러리

```python
import pandas as pd # 데이터 다루기 + 데이터셋 불러오기

import tensorflow as tf # 날씨 예측 인공지능 구현하기 위한 라이브러리

import numpy as np # 데이터 추가적으로 다루기 위한 라이브러리
```

데이터 셋 불러오기

```python
df = pd.read_csv(r'Dataset.csv',encoding='cp949') # 데이터 셋 불러오기
```

데이터 내용 확인

```python
df
```

데이터 전처리 - 불필요한 데이터 제거

```python
df = df.drop(columns=['지점'])
```

데이터 columns 확인 (불필요한 데이터 요소 제거가 되었는지 확인용)

```python
df.columns
```

```python
y = df.drop(columns=['지점명','일시'])
```

```python
y = y.fillna(0)
```

```python
y
```

정규화 처리(모든 데이터 -1 ~ 1사이 범위로 변경)

```python
y_mean = y.mean().to_numpy()
y_std = y.std().to_numpy()
y = (y-y.mean())/y.std()
```

정규화 처리 후 데이터 확인

```python
y
```

학습할 데이터 제작

```python
from IPython.display import Image
  
# 이미지 넣기
Image(url="weatherAL.png")
```

```python
trainY = []
trainX = []
for i in range(y.shape[0]-3):
    trainX.append(y[i:i+3].to_numpy())
    trainY.append(y[i+3:i+4].to_numpy())

trainX = np.array(trainX)
trainY = np.array(trainY)
```

인공지능 학습할 수 있게 데이터 구조 변경

```python
trainY = trainY.reshape((-1,19))
```

```python
trainX
```

인공지능 모델 구현(LSTM 이용한 날씨 예측)

```python
model = tf.keras.Sequential()

model.add(tf.keras.layers.LSTM(256,activation='tanh',input_shape=(3,19),return_sequences=True))
model.add(tf.keras.layers.LSTM(128,activation='tanh'))
model.add(tf.keras.layers.Dense(19,activation='tanh'))
```

모델 구조 확인용

```python
model.summary()
```

학습방법 설정

```python
model.compile(loss='mse',optimizer='adam') # loss='mse' 사용한 이유 : 회귀 문제이기 때문에
```

인공지능 학습 실행

```python
model.fit(trainX,trainY,epochs=50)
```

다음날 예측할 데이터 출력

```python
trainX[-1]
```

다음날 날씨 예측 후 DataFrame 생성

```python
result = pd.DataFrame(model.predict(trainX[-1:]).reshape((-1,19)))
# *y_std+y_mean
```

```python
result
```
