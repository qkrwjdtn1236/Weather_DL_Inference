```python
import pandas as pd

import tensorflow as tf

import numpy as np
```

```python
df = pd.read_csv(r'Dataset.csv',encoding='cp949')
```

```python
df
```

```python
df = df.drop(columns=['지점'])
```

```python
df['year'] = pd.DatetimeIndex(df['일시']).year
df['month'] = pd.DatetimeIndex(df['일시']).month
df['day'] = pd.DatetimeIndex(df['일시']).day
```

```python
df
```

```python
x = df[['지점명','month','day']]
```

```python
x
```

```python
x = pd.concat([pd.get_dummies(x['month']),pd.get_dummies(x['day'])],
          axis=1)
```

```python
x = pd.concat([pd.get_dummies(df['지점명']),x],
              axis=1)
```

```python
x
```

```python
df.columns
```

```python
y = df.drop(columns=['지점명','일시','year','month','day'])
```

```python
y = y.fillna(0)
```

```python
y
```

```python
y_mean = y.mean().to_numpy()
y_std = y.std().to_numpy()
y = (y-y.mean())/y.std()
```

```python
y
```

```python
#y_ = y*y_std+y_mean
```

```python
x.shape[1]
```

```python
y.shape
```

```python
trainY = []
trainX = []
for i in range(y.shape[0]-3):
    trainX.append(y[i:i+3].to_numpy())
    trainY.append(y[i+3:i+4].to_numpy())

trainX = np.array(trainX)
trainY = np.array(trainY)
```

```python
trainX.shape
```

```python
trainY.shape
```

```python
trainY = trainY.reshape((-1,19))
```

```python
trainX
```

```python
model = tf.keras.Sequential()

model.add(tf.keras.layers.LSTM(256,activation='tanh',input_shape=(3,19),return_sequences=True))
model.add(tf.keras.layers.LSTM(128,activation='tanh'))
model.add(tf.keras.layers.Dense(19,activation='tanh'))
```

```python
model.summary()
```

```python
model.compile(loss='mse',optimizer='adam')
```

```python
model.fit(trainX,trainY,epochs=50)
```

```python
trainX[-1]
```

```python
result = pd.DataFrame(model.predict(trainX[-1:]).reshape((-1,19)))
# *y_std+y_mean
```
